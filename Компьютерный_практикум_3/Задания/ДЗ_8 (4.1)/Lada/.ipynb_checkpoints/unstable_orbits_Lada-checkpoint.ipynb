# Неустойчивые орбиты вокруг точки L1

**Дано:**
- движение невесомого тела в системе из двух массивных тел, вращающихся относительно общего центра масс, описывается системой уравнений:

$\begin{cases}
\ddot{x} - 2\dot{y} = \frac{\partial{U}}{\partial{x}} \\
\ddot{y} + 2\dot{x} = \frac{\partial{U}}{\partial{y}} \\
\ddot{z} = \frac{\partial{U}}{\partial{z}} \\
U = \frac{1}{2} (x^2 + y^2) + \frac{\mu_1}{r_1} + \frac{\mu_2}{r_2} \\
r^2_1 = (x + \mu_2)^2 + y^2 + z^2 \\
r^2_2 = (x - \mu_1)^2 + y^2 + z^2
\end{cases}$

- константы модели для системы Солнце-Земля:
    - $\mu_2 = 3.001348389698916e-06$

# Пункт 1 (+проверка)

````python
import numpy as np
from numba import njit, cfunc, prange
from time import time
import matplotlib.pyplot as plt
````

````python
jkw = dict(cache=True)


@cfunc('f8[:](f8, f8[:], f8[:])', **jkw)
def crtbp_ode(t, s, mc):
    """
    функция f в системе s'=f(t,s) в систему ДУ для задачи ОКЗТТ с набором постоянных mc
    """
    x, y, z, vx, vy, vz = s
    mu2 = mc[0]
    mu1 = 1. - mu2

    r1 = ((x + mu2)**2 + y**2 + z**2)**0.5
    r2 = ((x - mu1)**2 + y**2 + z**2)**0.5

    ax = 2*vy + x - (mu1 * (x + mu2)/r1**3 + mu2 * (x - mu1)/r2**3)
    c = (mu1 / r1**3 + mu2 / r2**3)
    ay = -2*vx + y - c * y
    az = -c * z

    return np.array([vx, vy, vz, ax, ay, az])


@njit(**jkw)
def rk_step_Abc(t, s, h, A, b, c, mc):
    """
    шаг численного решения системы ДУ ОКЗТТ (s'=crtbp_ode(t,s,mc))
    с размером h методом Рунге-Кутта с таблицей Бутчера (A,b,c)
    """
    N = b.shape[0]
    k = np.empty((N, 6), dtype=np.float64)
    s_sum = np.zeros(6, dtype=np.float64)
    for i in range(N):
        k_sum = np.zeros(6, dtype=np.float64)
        for j in range(i):
            k_sum += A[i,j]*k[j]
        k[i] = crtbp_ode(t + h * c[i], s + h * k_sum, mc)
        s_sum += h * b[i] * k[i]
    return s + s_sum


@njit(**jkw)
def rk_nsteps_Abc(t, s, h, tab, mc, n):
    """
    функция расчёта численного решения ДУ ОКЗТТ в n шагов начиная от t0 = t, s0 = s
    с шагом h, таблицей бутчера tab
    """
    A, b, c = tab
    arr = np.empty((n + 1, s.shape[0] + 1))
    arr[:, 0] = t + h * np.arange(n + 1)
    arr[0, 1:] = s
    for i in range(n):
        arr[i + 1, 1:] = rk_step_Abc(
                                  arr[i, 0],
                                  arr[i, 1:],
                                  h, A, b, c, mc)  
    return arr


@njit(**jkw)
def rk_nsteps_planes_Abc(s, h, tab, mc, n, pl):
    """
    функция расчёта численного решения ДУ ОКЗТТ в n шагов начиная от t0 = t, s0 = s
    с шагом h, таблицей бутчера tab, огранниченная по x pl 
    """
    A, b, c = tab
    arr = np.empty((n + 1, s.shape[0] + 1), dtype=np.float64)
    arr[:, 0] = h * np.arange(n + 1)
    arr[0, 1:] = s
    
    i = 0
    for i in range(n):
        arr[i + 1, 1:] = rk_step_Abc(
                                  arr[i, 0],
                                  arr[i, 1:],
                                  h, A, b, c, mc) 
        x = arr[i + 1, 1]
        if x < pl[0] or x > pl[1]:
            break

    return arr[:i + 2]
````

````python
@njit
def get_plane(vy, s, h, tab, mc, steps, pl):
    """
    "знаковая" функция результата, где оказался аппарат после моделирования относительно
    xL1 - слева -1 или справа 1
    при начальной состоянии [*s[0:2],0,vy,0] - помогает работе метода бисекции
    """
    s0 = s.copy()
    s0[4] = vy
    arr = rk_nsteps_planes_Abc(s0,h,tab,mc,steps,pl)
    x = arr[-1,1]
    xmean = np.mean(pl)
    return -1 if x < xmean else 1


from scipy.optimize import bisect

def get_v0(s3, h, tab, mc, steps, pl):
    """
    функция метода бисекции для определения начальной скорости аппарата для долгого пребывания вокруг xL1
    через поиски точки разрыва get_planе при начальной положении s3
    """
    s0 = np.zeros(6)
    s0[:3] = s3
    v_star = bisect(get_plane, -0.1, 0.1, args=(s0, h, tab, mc, steps, pl), xtol=1e-16)
    return v_star

def funcGrid(borders, N, h, tab, mc, steps, printable=False, count_time=True):
    """
    функция расчёта значения начальной скорости аппарата по решетке начальных состояний,
    ограниченной borders (решётка N x N)
    """
    x_min, x_max, z_min, z_max, x_left, x_right = borders
    xGrid = np.linspace(x_min, x_max, N)
    zGrid = np.linspace(z_min, z_max, N)
    pl = np.array([x_left, x_right])
    res = np.empty((N,N), dtype=np.float64)
    t1 = time()
    for i in range(N):
        for j in range(N):
            s = np.array([xGrid[i], 0, zGrid[j]])
            res[i,j] = get_v0(s, h, tab, mc, steps, pl)
    t2 = time()
    np.save(f'v0_grid_{N}.npy', res)
    if count_time:
        print(f'{round(t2 - t1,1)} seconds for N = {N}')
    if printable:
        print(res)
    
````

````python
A = np.array([[0,         0,          0,          0,          0,              0,    0],
              [1/5,       0,          0,          0,          0,              0,    0],
              [3/40,      9/40,       0,          0,          0,              0,    0],
              [44/45,     -56/15,     32/9,       0,          0,              0,    0],
              [19372/6561,-25360/2187,64448/6561, -212/729,   0,              0,    0],
              [9017/3168, -355/33,    46732/5247, 49/176,     -5103/18656,    0,    0],
              [35/384,    0,          500/1113,   125/192,    -2187/6784,     11/84,0]])
c = np.array([0, 1/5, 3/10, 4/5, 8/9, 1, 1])
b = np.array([35/384, 0, 500/1113, 125/192, -2187/6784, 11/84, 0])
````

# Данные для запусков

````python
xL1 = 0.9900289479871318
mu2 = 3.001348389698916e-06
mu1 = 1 - mu2
R = 149600000
mc = np.array([mu2])
d_p = 1400000. / R
d_g = 1000000. / R
h = 0.01721420632103996
borders = (xL1 - d_g / 2, xL1 + d_g / 2, 0., d_g, xL1 - d_p, xL1 + d_p)
steps = 100000
````

# Запуск экспериментов

````python
# "прогреваем" кэш
funcGrid(borders, 2, h, (A,b,c), mc, steps, printable=True)
````

# Пункт 8 (графики)

````python
fig, ax = plt.subplots(2, 4, figsize=(20,10))
for k, N in enumerate([40, 80, 160, 320]):
    x_min, x_max, z_min, z_max, _, _ = borders
    x_reg = np.linspace(x_min, x_max, N)
    z_reg = np.linspace(z_min, z_max, N)
    data = np.load(f'v0_grid_{N}.npy')
    im1 = ax[0,k].imshow(data.T, extent=[(x_min - xL1) * R, (x_max - xL1) * R, z_min * R, z_max * R])
    ax[0,k].set_title(f'v0 для решётки {N}')
    ax[0,k].set_xlabel('x')
    ax[0,k].set_ylabel('z')
    J = np.empty((N,N))
    for i in range(N):
        for j in range(N):
            r1 = np.sqrt((x_reg[i] + mu2) ** 2 + z_reg[j] ** 2)
            r2 = np.sqrt((x_reg[i] - mu1) ** 2 + z_reg[j] ** 2)
            U = x_reg[i] ** 2 / 2 + mu1 / r1 + mu2 / r2
            J[i,j] = 2 * U - data[i,j] ** 2
    
    im2 = ax[1,k].imshow(J.T, extent=[(x_min - xL1) * R, (x_max - xL1) * R, z_min * R, z_max * R])
    ax[1,k].set_title(f'J для решётки {N}')
    ax[1,k].set_xlabel('x')
    ax[1,k].set_ylabel('z')
plt.colorbar(im1, ax=ax[0,-1]);
plt.colorbar(im2, ax=ax[1,-1]);
````
