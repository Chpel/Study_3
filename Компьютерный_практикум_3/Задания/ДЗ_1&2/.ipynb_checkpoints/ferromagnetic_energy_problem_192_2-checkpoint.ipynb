# Энергия ферромагнетика

### Условие задачи

**Дано:**
- цепочка молекул, расположенных в узлах кристаллической решетки, длины $L$
- каждая молекула обладает спином +1 или -1
- межмолекулярное взаимодействие описывается константами $J_{ij} = 1$
- модель Изинга


**Требуется:**
- согласно модели Изинга рассчитать среднюю энергию $<E>$ для указанной цепочки молекул при:
    - длине цепочки $L \in [2, 3, ..., 31]$
    - температурах $kT \in [1, 1.1, ..., 5.0]$
- вывести время расчета каждой итерации по $L$ или по $kT$
- отобразить цветовую карту:
    - ось абсцисс - $L$,
    - ось ординат - $kT$,
    - цветом отобразить нормированное значение средней энергии $\langle E \rangle / L$,
    - подписать оси,
    - отобразить цветовую шкалу (colorbar),
    - засечки должны соответствовать значениям $L, kT$.
    

**Описание:**

Модель Изинга является моделью магнетика. Пусть этот магнетик состоит из молекул, расположенных в узлах регулярной решетки. Пусть всего таких узлов будет $N$ штук, с индексами $i=1,\ldots, N$.

Предположим, что каждая молекула может быть представлена в виде магнитной стрелки, которая всегда либо направлена вдоль некоторой заданной оси, либо в противоположном направлении. То есть каждая молекула $i$ имеет две конфигурации, которые можно описывать с помощью "спиновой" переменной $\sigma_i$. Эта переменная принимает значение $+1$ (параллельно оси, спин направлен вверх) и $-1$ (антипараллельно оси, спин направлен вниз).

Пусть $\sigma = \{\sigma_1, \sigma_2, \ldots, \sigma_N\}$ обозначает набор значений всех $N$ спинов. Имеется $2^N$ различных наборов $\sigma$, и каждый из них описывает некоторое состояние системы. 

Гамильтониан системы  состоит из двух частей: первая $E_0$ включает вклад межмолекулярных сил внутри магнетика, а вторая $E_1(\sigma)$ вклад от взаимодействий каждого спина с внешним магнитным полем (здесь считается нулевым). 
$$H(\sigma)=E_0(\sigma)+E_1(\sigma) \label{hamiltonian}$$

В любой физической системе мы предполагаем все взаимодействия инвариантными по отношению к обращению времени, что озачает инвариантность $E$ при изменении знаков всех полей и намагниченностей. Энергия должна быть четной функцией от $\sigma$:
$$E_0(\sigma_1,\ldots, \sigma_N)=E_0(-\sigma_1,\ldots, -\sigma_N)$$

Энергия системы при нулевом внешнем магнитном поле равна сумме произведений соседних спинов на константы взаимодействия $J_{ij}$
$$E(\sigma) = -\sum_{\langle i,j\rangle} J_{ij}\sigma _i\sigma _j $$

Вероятность находиться в состоянии $\sigma$
$$P_{\beta}(\sigma)=\frac{e^{-\beta E(\sigma)}}{Z_{\beta}},$$
	где $Z_{\beta} = \sum_{\sigma} e^{-\beta E(\sigma)}-$ статистическая сумма, $\beta = \frac{1}{k_bT}-$ обратная температура, $k_b-$ константа Больцмана.
	
Средняя энергия системы $$\langle E \rangle = \frac{1}{Z}\sum_{\{\sigma \}} E(\sigma)e^{-\frac{E(\sigma)}{kT}}$$

````python
import numpy as np
from time import time
````

**Прототип**

````python
def E(sigma: np.ndarray, J: np.ndarray) -> float:
    """
    Программа расчитывает энергию текущей конфигурации: для каждой пары спинов (включая концы конформаций) 
    и их константы взаимодействия (здесь она может быть отлична от единицы)
    рассчитывается энергия и суммируется со знаком минус
    
    Параметры:
    sigma : конфигурация спинов, массив numpy
    J : массив констант взаимодействия для каждой пары соседний спинов, массив numpy
    
    Возвращает:
    Значение энергии в данной конфигурации
    """
    return -np.sum(J * sigma * np.roll(sigma, -1))

def Emean(L: int, kT: np.ndarray, J0=None, OpenQ=True) -> float:
    """
    Программа перебирает всевозможные конфигурации, начиная от "всех единиц", рассчитывает для них энергию:
    - энергия необходима для расчёта неприведённой вероятности данного состояния, которое является слагаемым статсуммы
    - в так же для расчёта средней энергии
    - зная от четности функции энергии конфигурации, мы может считать в два раз меньше конфигураций (например, только те, где первый спин с одним значением)
    - изменение значения первого спина означает остановку работы программы
    - в программе так же есть функция автозаполнения массива констант - в зависимости от граничных условий
    
    Параметры:
    L : длина цепочки спинов, целое число
    kT : температура, вещественное число
    J0 : массив констант взаимодействия, массив numpy
    OpenQ : флаг открытых граничных условий
    
    Возвращает:
    Значение средней энергии цепочки заданной длины
    """
    sigma = np.ones(L)
    Z = 0
    Em = 0
    if J0 == None:
        J = np.ones(L - 1)
    else:
        J = J0
    if OpenQ:
        J = np.append(J, 0)
    elif J0 == None:
        J = np.append(J, 1)
    while True:
        Es = E(sigma, J)
        Ps = np.exp(-Es / kT)
        Z += Ps
        Em += Es * Ps
        for j in range(L):
            if sigma[L - 1 - j] == -1:
                sigma[L - 1 - j] = 1
                continue
            else:
                sigma[L - 1 - j] = -1
                break
        if np.all(sigma == 1):
            break
    return Em / Z 
````

**Проверочная функция для отслеживания правильности кода**

````python
def Emean_check(L, J, kT):
    return - (L - 1) * J * np.tanh(J / kT)
Emean_check(4, 1, 5)
````

**Проверка работы кода**

````python
L = np.arange(2,32)
kT = np.arange(1.0, 5.0 + 0.1, 0.1)
E_mean = np.vectorize(Emean, excluded=['J0', 'OpenQ'])
E_data = np.array([[]])
for l_0 in [2,3,4,5]:
    %time E_data_0 = E_mean(l_0, kT)
    if E_data.size == 0:
        E_data = np.expand_dims(E_data_0, axis=0)
    else:
        E_data = np.append(E_data, np.expand_dims(E_data_0, axis=0), axis=0)
    print(l_0, E_data_0)
````

**Компиляция в njit**

````python
import matplotlib.pyplot as plt
from numba import jit, njit, prange
````

````python
@njit(boundscheck=False)

def E_njit(sigma: np.ndarray) -> float:
    """
    Оптимизированная программа расчитывает энергию текущей конфигурации: для каждой пары спинов (включая концы конформаций)
    рассчитывается энергия и суммируется со знаком минус
    
    - система контроля выхода за границы массива отключена для ускорения работы (алгоритм задан так, что никогда не выйдет за пределы массива)
    - массив констант опущен для ускорения программы и соблюдения условия (Jij = 1)
    - так же алгоритм расчёта был оптимизирован - работа проводится над одним и тем же массивом
    
    Параметры:
    sigma : конфигурация спинов, массив numpy
    
    Возвращает:
    Значение энергии в данной конфигурации
    """
    E = 0
    L = sigma.size
    for i in range(L):
        E -= sigma[i] * sigma[(i+1) % L]
    return E
````

````python
@njit(boundscheck=False)

def Emean_njit(L: int, kT: np.ndarray) -> float:
    """
    Оптимизированная программа перебирает всевозможные конфигурации, начиная от "всех единиц", рассчитывает для них энергию:
    - система контроля выхода за границы массива отключена для ускорения работы (алгоритм задан так, что никогда не выйдет за пределы массива)
    - энергия необходима для расчёта неприведённой вероятности данного состояния, которое является слагаемым статсуммы
    - а так же для расчёта средней энергии
    - зная от четности функции энергии конфигурации, мы может считать в два раз меньше конфигураций (например, только те, где первый спин с одним значением)
    - изменение значения первого спина означает остановку работы программы
    - программа отошла от концепции вариации констант взаимодействия и граничных условий - для ускорения работы
    
    Параметры:
    L : длина цепочки спинов, целое число
    kT : температура, вещественное число
    
    Возвращает:
    Значение средней энергии цепочки заданной длины
    """
    sigma = np.ones(L)
    Z = np.zeros(kT.size)
    Em = np.zeros(kT.size)
    while True:
        Es = E_njit(sigma)
        Ps = np.exp(-Es / kT)
        Z += Ps
        Em += Es * Ps
        for j in range(L):
            if sigma[L - 1 - j] == -1:
                sigma[L - 1 - j] = 1
                continue
            else:
                sigma[L - 1 - j] = -1
                break
        if sigma[0] == -1:
            break
    return Em / Z 
````

````python
L=25
L_Full = 31
E_data = np.empty((L_Full - 1, kT.size))
for l_0 in range(2, L + 1):
    print(l_0, end=' ')
    %time E_data_0 = Emean_njit(l_0, kT)
    E_data[l_0 - 2] = E_data_0 / l_0
````

````python
L=L_Full
for l_0 in range(26, L + 1):
    print(l_0, end=' ')
    %time E_data_0 = Emean_njit(l_0, kT)
    E_data[l_0 - 2] = E_data_0 / l_0
````

````python
for l_0 in range(2,31+1):
    print(l_0, E_data[l_0 - 2])
````

````python
fig, ax = plt.subplots(figsize=(20,10))
L_all = np.arange(2, L+1,1)
im = ax.pcolormesh(kT, L_all, E_data, shading='nearest')
fig.colorbar(im, ax=ax)
ax.set_yticks(np.arange(2, L+1,2))
ax.set_yticks(np.arange(2, L+1,1), minor=True)
ax.set_xticks(np.arange(1.0, 5.0+0.2, 0.2))
ax.set_xticks(np.arange(1.0, 5.0+0.1, 0.1), minor=True)
plt.xlabel('kT')
plt.ylabel('L')
plt.show()
````
