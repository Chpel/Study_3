### Подготовка данных

```python
from torchvision.io import read_image, ImageReadMode

class MyDataset(torch.utils.data.Dataset):
    def __init__(self, data_dir, transform):
        self.img_dir = Path(data_dir)
        self.transform = transform
        self.classes = os.listdir(self.img_dir)
        self.k_classes = len(self.classes)
        self.items = os.listdir(self.img_dir / self.classes[0])
        self.k_items = len(self.items)
        pass
    
    def __getitem__(self, idx):
        class_num = idx // self.k_items
        item_num = (idx % self.k_items)
        img = read_image(str(self.img_dir / self.classes[class_num] / self.items[item_num]), ImageReadMode.RGB).float()
        
        if self.transform is not None:
            img = self.transform(img)
        
        return img, class_num
    
    def __len__(self):
        return self.k_classes * self.k_items
```

```python
def set_random_seed(seed):
    torch.backends.cudnn.deterministic = True
    torch.manual_seed(seed)
    torch.cuda.manual_seed(seed)
    np.random.seed(seed)
    random.seed(seed)
```

```python
train_transform = transforms.Compose([
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
]
)
val_transform = transforms.Compose([
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
]
)
# YOU CAN DEFINE AUGMENTATIONS HERE

set_random_seed(100)

train_dataset = MyDataset("./dataset/dataset/dataset/train", transform=train_transform)
val_dataset = MyDataset("./dataset/dataset/dataset/val", transform=val_transform)

train_dataloader = torch.utils.data.DataLoader(train_dataset, batch_size=100, shuffle=True)
val_dataloader = torch.utils.data.DataLoader(val_dataset, batch_size=100, shuffle=True)
```

### Вспомогательные функции, реализация модели

```python
def train_one_epoch(model, train_dataloader, criterion, optimizer, device="cuda:0"):
    model.train()
    # YOUR CODE
    # TRAIN YOUR MODEL HERE
    with tqdm(total=len(train_dataloader), file=sys.stdout) as prbar:
        for images, labels in train_dataloader:
            # Move Batch to GPU
            images = images.to(device)
            labels = labels.to(device)
            predicted = model(images)
            loss = criterion(predicted, labels)
            # Update weights
            loss.backward()
            optimizer.step()
            optimizer.zero_grad()
            # Update descirption for tqdm
            accuracy = (predicted.argmax(1) == labels).float().mean()
            prbar.set_description(
                f"Loss: {round(loss.item(), 4)} "
                f"Accuracy: {round(accuracy.item() * 100, 4)}"
            )
            prbar.update(1)

def predict(model, val_dataloader, criterion, device="cuda:0"):
    model.to(device).eval()
    # YOUR CODE
    # PREDICT FOR EVERY ELEMENT OF THE VAL DATALOADER AND RETURN CORRESPONDING LISTS
    predicted_classes = np.array([])
    true_classes = np.array([])
    all_losses = np.array([])
    with tqdm(total=len(val_dataloader), file=sys.stdout) as prbar:
        for images, labels in val_dataloader:
            images = images.to(device)
            labels = labels.to(device)
            predicted = model(images)
            loss = criterion(predicted, labels)
            accuracy = (predicted.argmax(1) == labels).float().mean()
            prbar.set_description(
                f"Loss: {round(loss.item(), 4)} "
                f"Accuracy: {round(accuracy.item() * 100, 4)}"
            )
            prbar.update(1)
            all_losses = np.append(all_losses, loss.item())
            predicted_classes = np.append(predicted_classes, predicted.argmax(1).cpu().detach().numpy())
            true_classes = np.append(true_classes, labels.cpu().detach().numpy())
    return all_losses, predicted_classes, true_classes


def train(model, train_dataloader, val_dataloader, criterion, optimizer, device="cuda:0", n_epochs=10, scheduler=None):
    model.to(device)
    for epoch in range(n_epochs):
        # YOUR CODE
        # Train, evaluate, print accuracy, make a step of scheduler or whatever you want...
        print(f"Epoch {epoch}:")
        train_one_epoch(model, train_dataloader, criterion, optimizer, device)
        train_loss, train_pred_y, train_true_y = predict(model, train_dataloader, criterion, device)
        val_loss, val_pred_y, val_true_y = predict(model, val_dataloader, criterion, device)
        print(f"train_loss = {np.sum(train_loss) / len(train_loss)}")
        print(f"train_accuracy = {accuracy_score(train_pred_y, train_true_y)}")
        print(f"val_loss = {np.sum(val_loss) / len(val_loss)}")
        print(f"val_accuracy = {accuracy_score(val_pred_y, val_true_y)}")
        if scheduler is not None:
          scheduler.step()
```

### Обучение модели, запуски экспериментов

```python
import torch.cuda

model = nn.Sequential(
    nn.Conv2d(3, 8, 3, padding=1),
    nn.BatchNorm2d(8),
    nn.ReLU(),
    nn.Conv2d(8, 8, 3, padding=1),
    nn.BatchNorm2d(8),
    nn.ReLU(),
    nn.MaxPool2d(2, 2),
    nn.Dropout(0.2),

    nn.Conv2d(8, 16, 3, padding=1),
    nn.BatchNorm2d(16),
    nn.ReLU(),
    nn.Conv2d(16, 16, 3, padding=1),
    nn.BatchNorm2d(16),
    nn.ReLU(),
    nn.MaxPool2d(2, 2),
    nn.Dropout(0.2),

    nn.Conv2d(16, 32, 3, padding=1),
    nn.BatchNorm2d(32),
    nn.ReLU(),
    nn.Conv2d(32, 32, 3, padding=1),
    nn.BatchNorm2d(32),
    nn.ReLU(),
    nn.MaxPool2d(2, 2),
    nn.Dropout(0.2),

    nn.Conv2d(32, 64, 3, padding=1),
    nn.BatchNorm2d(64),
    nn.ReLU(),
    nn.Conv2d(64, 64, 3, padding=1),
    nn.BatchNorm2d(64),
    nn.ReLU(),
    nn.MaxPool2d(2, 2),
    nn.Dropout(0.2),

    nn.Flatten(),
    nn.Linear(1024, 256),
    nn.BatchNorm1d(256),
    nn.ReLU(),
    nn.Dropout(0.3),
    nn.Linear(256, 200),
)

criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), 2e-3)
scheduler = StepLR(optimizer, 5, 0.5)
n_epochs = 20
device = "cuda:0" if torch.cuda.is_available() else "cpu"
```
