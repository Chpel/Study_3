```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def implicit_euler_solve(u0, T, dt):
    """Решает $du/dt = 1/u$ на $0 < t < T$ с $u(t=0) = u0$ при помощи неявного метода Эйлера."""
    num_steps = int(T/dt)
    tt = np.arange(num_steps+1)*dt
    y = np.empty((num_steps+1,2))
    y[0] = [u0, u0]
    sign = np.array([1,-1])
    for k in range(num_steps):
        y[k+1] = (y[k] + sign * np.sqrt(np.power(y[k], 2) + 4 * dt)) / 2
    return tt, y.T
```

```python
x, y = implicit_euler_solve(1, 5, 0.1)
```

```python
plt.plot(x, y[0], label='num1')
plt.plot(x, y[1], label='num2')
plt.plot(x, np.sqrt(2 * np.array(x)) + 1, label='true1')
plt.plot(x, -np.sqrt(2 * np.array(x)) + 1, label='true2')
plt.legend()
```
