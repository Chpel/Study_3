# VI.1 (Пчелинцев -  нечетное)

Формула преобразования из равномерного в экспоненциальное:

$X_i = -\frac{1}{\lambda} \log(1-U_i)$, где $U_i$ - элемент равномерной выборки, а $X_i$ - экспоненциальной

```python
def uni_to_exp(U, lam):
    return -1/lam * np.log(1-U)
```

```python
lam = 1
for n in [100, 200, 500, 1000, 2000, 5000]:
    U = np.random.rand(n)
    X = uni_to_exp(U, lam)
    X_true = np.random.exponential(1/lam, n)
    plt.hist([X, X_true], label=['exp_calc', 'exp_true'], density=True)
    plt.title(fr'n = {n}, $\lambda$ = {lam}')
    plt.legend();
    plt.show()
    print(kstest(X, expon(scale=1.0/lam).cdf, mode='exact'))
```

Заметно, что выборки с ростом их размеров становятся более похожими друг на друга

# VI.3

```python
data = np.load('data_89.npz')
A = data['A4']
```

```python
A.shape
```

$I = \int \prod_{i=1}^{n}dx_i \frac{\exp{(-x^T A x)}}{1+x_1^2+\dots+x_n^2}$

Плотностью вероятности выступит многомерное нормальное распределение:

$w(x) = \frac{1}{\sqrt{(2\pi)^n det \sum}} \exp{(-\frac{1}{2}(x-\mu)^T \sum^{-1} (x-\mu))}$

В нашем случае, $\sum = (2A)^{-1},\ \ \mu = 0$

Тогда $w(x) = \frac{1}{\sqrt{(2\pi)^n det (2A)^{-1}}} \exp{(-x^T A x)}$

$I = \int \prod_{i=1}^{n}dx_i \frac{\sqrt{(2\pi)^n det (2A)^{-1}}}{1+x_1^2+\dots+x_n^2} w(x)$

Получается, все вектора будут генерироваться по многомерному нормальному закону

```python
detE = np.linalg.det(np.linalg.inv(2*A))
def I(r):
    d = r.shape[-1]
    ari_wri = []
    for r_j in r:
        ari_wri.append(np.sqrt((2*np.pi)**d * detE)/(1 + (r_j ** 2).sum()))
    return np.mean(np.asarray(ari_wri))
```

```python
gen = np.random.default_rng(1234)
d = A.shape[0]
for N in [10000, 20000, 50000, 100000]:
    r = gen.multivariate_normal(np.zeros((d, )), 2*A, N)
    print("N: %6d    I: %.6f" % (N, I(r)))
```

# VII.1

```python
import autograd.numpy as np

def f_x_a(x, a):
    return 1/(np.exp(a[0] + np.dot(a[1:], x)) + 1)

def leastsquares(a, x, y):
    s = 0
    for i in range(len(x)):
        s += (f_x_a(x[i], a) - y[i]) ** 2
    return s
```

```python
x = data['A2']
y = data['y2']
```

```python
result = minimize(leastsquares, np.zeros(x.shape[-1] + 1), (x, y))
```

```python
result.x
```

```python
leastsquares(result.x, x, y)
```

# VII.2

```python
def grad_manual(f, x0, h):
    d = x0.shape[0]
    gr = np.zeros(d)
    for i in range(d):
        x = x0.copy()
        x[i] += h
        gr[i] = (f(x) - f(x0)) / h
    return gr
```

```python
gen = np.random.default_rng(1234)
```

```python
gamma = 0.004
a0 = gen.uniform(size = (x.shape[-1] + 1,))
```

```python
a = a0.copy()
for _ in tqdm(range(10000)):
    a -= gamma * grad_manual(lambda a: leastsquares(a, x, y), a, 0.00001)
print(leastsquares(a, x, y))
```

```python
a = a0.copy()
gr = grad(leastsquares)
for _ in tqdm(range(10000)):
    a -= gamma * gr(a, x, y)
print(leastsquares(a, x, y))
```
