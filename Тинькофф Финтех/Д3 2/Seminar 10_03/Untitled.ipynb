1. Ð ÐµÐ°Ð»Ð¸Ð·Ð¾Ð²Ð°Ñ‚ÑŒ Ñ„ÑƒÐ½ÐºÑ†Ð¸ÑŽ feature_line(var, target, nbins)

    Ð’ ÐºÐ°Ñ‡ÐµÑÑ‚Ð²Ðµ Ð°Ñ€Ð³ÑƒÐ¼ÐµÐ½Ñ‚Ð¾Ð² Ñ„ÑƒÐ½ÐºÑ†Ð¸Ñ Ð´Ð¾Ð»Ð¶Ð½Ð° Ð¿Ñ€Ð¸Ð½Ð¸Ð¼Ð°Ñ‚ÑŒ:
     - var: Ð²ÐµÐºÑ‚Ð¾Ñ€ Ð·Ð½Ð°Ñ‡ÐµÐ½Ð¸Ð¹ Ð¿Ñ€Ð¸Ð·Ð½Ð°ÐºÐ°
     - target: Ð²ÐµÐºÑ‚Ð¾Ñ€ Ð·Ð½Ð°Ñ‡ÐµÐ½Ð¸Ð¹ Ñ†ÐµÐ»ÐµÐ²Ð¾Ð¹ Ð¿ÐµÑ€ÐµÐ¼ÐµÐ½Ð½Ð¾Ð¹
     - nbins: ÐºÐ¾Ð»Ð¸Ñ‡ÐµÑÑ‚Ð²Ð¾ Ð±Ð°ÐºÐµÑ‚Ð¾Ð²

    Ð¸ Ð²Ð¾Ð·Ð²Ñ€Ð°Ñ‰Ð°Ñ‚ÑŒ:
     - Ð³Ñ€Ð°Ñ„Ð¸Ðº Ð»Ð¸Ð½ÐµÐ¹Ð½Ð¾ÑÑ‚Ð¸ Ð¿Ñ€Ð¸Ð·Ð½Ð°ÐºÐ°
     - ÐºÐ¾Ð»Ð¸Ñ‡ÐµÑÑ‚Ð²Ð¾ ÑƒÐ´Ð°Ð»ÐµÐ½Ð½Ñ‹Ñ… Ð½Ð°Ð±Ð»ÑŽÐ´ÐµÐ½Ð¸Ð¹
     - Ð·Ð½Ð°Ñ‡ÐµÐ½Ð¸Ðµ Ð¼ÐµÑ‚Ñ€Ð¸ÐºÐ¸ R2
     - Ð³Ð¸ÑÑ‚Ð¾Ð³Ñ€Ð°Ð¼Ð¼Ñƒ Ñ€Ð°ÑÐ¿Ñ€ÐµÐ´ÐµÐ»ÐµÐ½Ð¸Ñ Ð¿Ñ€Ð¸Ð·Ð½Ð°ÐºÐ°
 
 
2. ÐŸÑ€Ð¾Ð²ÐµÑ€Ð¸Ñ‚ÑŒ Ð»Ð¸Ð½ÐµÐ¹Ð½Ð¾ÑÑ‚ÑŒ Ð¿Ñ€Ð¸Ð·Ð½Ð°ÐºÐ¾Ð² Ñ Ð¿Ð¾Ð¼Ð¾Ñ‰ÑŒÑŽ Ñ„ÑƒÐ½ÐºÑ†Ð¸Ð¸ feature_line()

```python
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression

def feature_line(var, target, nbins):
    N = len(target)
    idx = np.argsort(var)
    bin_len = N // nbins
    bin_last_len = bin_len + N % nbins
    x = []
    y = []
    y_erru = []
    y_errd = []
    fig, ax = plt.subplots(figsize=(15,8), ncols=2)
    for i in range(nbins):
        start = i * bin_len
        if i != nbins - 1:
            end = start + bin_len
        else:
            end = start + bin_last_len
        idx_bucket = idx[start:end]
        X_bucket = var.iloc[idx_bucket]
        y_bucket = target.iloc[idx_bucket]
        x.append(X_bucket.mean())
        y.append(y_bucket.mean())
        y_errd.append(abs(y_bucket.quantile(.25) - y_bucket.mean()))
        y_erru.append(abs(y_bucket.quantile(.75) - y_bucket.mean()))
        ax[1].hist(X_bucket, 20, label=f'bucket_{i}')
    ax[0].errorbar(x,y,yerr=[y_errd, y_erru], fmt='o--', capsize=5)
    ax[0].set_xlabel("feature")
    ax[0].set_ylabel("target")
    ax[1].legend()
    model = LinearRegression()
    model.fit(np.array(var).reshape(-1,1),target)
    print(f"R2_train: {round(r2_score(target, model.predict(np.array(var).reshape(-1,1))),4)}\n")
    ax[0].plot(x, model.predict(np.array(x).reshape(-1,1)))
    plt.show()
```

```python
def get_residuals(Y,Pred):
    '''ÐŸÐ¾Ð´ÑÑ‡ÐµÑ‚ Ð½ÐµÐ²ÑÐ·ÐºÐ¸ Y - Ð¸ÑÑ‚Ð¸Ð½Ð½Ð¾Ðµ Ð·Ð½Ð°Ñ‡ÐµÐ½Ð¸Ðµ Ñ‚Ð°Ñ€Ð³ÐµÑ‚Ð°, Pred - Ð¿Ñ€ÐµÐ´ÑÐºÐ°Ð·Ð°Ð½Ð½Ð¾Ðµ'''
    return Y - Pred
```

```python
def check_homoscedacity(predictor,target,f_name):
    simple_model = LinearRegression()
    simple_model.fit(predictor.reshape(-1,1),target)
    predicts = simple_model.predict(predictor.reshape(-1,1))
    return (hv.Scatter(zip(predicts,get_residuals(target,predicts)),\
              kdims = ['Y_estimate'],vdims=['Residual'], 
                       label = f'ÐŸÑ€Ð¾Ð²ÐµÑ€ÐºÐ° Ð³Ð¾Ð¼Ð¾ÑÐºÐµÐ´Ð°ÑÑ‚Ð¸Ñ‡Ð½Ð¾ÑÑ‚Ð¸ Ð¿Ñ€Ð¸Ð·Ð½Ð°ÐºÐ° {f_name}')
            .opts(plot=dict(width=1200,height=900)))
```

Ð’Ð¸Ð·ÑƒÐ°Ð»Ð¸Ð·Ð¸Ñ€ÑƒÐµÐ¼ Ð·Ð°Ð²Ð¸ÑÐ¸Ð¼Ð¾ÑÑ‚ÑŒ Ð´Ð»Ñ Ð¿Ñ€ÐµÐ´ÑÐºÐ°Ð·Ð°Ð½Ð¸Ñ Ð»Ð¸Ð½ÐµÐ¹Ð½Ð¾Ð¹ Ñ€ÐµÐ³Ñ€ÐµÑÑÐ¸Ð¸ Ð¾Ñ‚ Ð½ÐµÐ¿Ñ€ÐµÐ¾Ð±Ñ€Ð°Ð·Ð¾Ð²Ð°Ð½Ð½Ð¾Ð³Ð¾ Ð¿Ñ€Ð¸Ð·Ð½Ð°ÐºÐ° in_total_amt Ð¸ Ñ†ÐµÐ»ÐµÐ²Ð¾Ð¹ Ð¿ÐµÑ€ÐµÐ¼ÐµÐ½Ð½Ð¾Ð¹:

```python
def make_bucket(df,feature, n =  100):
    '''Ñ„ÑƒÐ½ÐºÑ†Ð¸Ñ, Ð±ÑŒÑŽÑ‰Ð°Ñ Ð½Ð° Ð±Ð°ÐºÐµÑ‚Ñ‹(Ð¿Ð¾ ÑƒÐ¼Ð¾Ð»Ñ‡Ð°Ð½Ð¸ÑŽ 100 Ñ‚Ð¾Ñ‡ÐµÐº)'''
    return df.assign(bucket = np.ceil(df[feature].rank(pct = True) * n))
```

```python
def plot_homoscedacity(predictor, target, predictor_name='feature'):
    df_tmp = pd.DataFrame({'predictor':predictor, 'target':target})
    return df_tmp.pipe(make_bucket,'predictor',200)\
                 .groupby(by = ['bucket']).mean()\
                 .pipe(lambda x:check_homoscedacity(np.array(x.predictor),np.array(x.target),predictor_name)) 
```
